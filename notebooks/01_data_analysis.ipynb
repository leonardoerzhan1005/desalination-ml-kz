```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Настройка графиков
sns.set(style="whitegrid")

# Загрузка данных
df = pd.read_csv("../data/sensor_data.csv")
df.head()

```

```python
df.hist(figsize=(12, 8), bins=20)
plt.suptitle("Распределение параметров", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Корреляционная матрица")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x="salinity", y="target_output_pressure")
plt.title("Соленость vs Выходное давление")
plt.show()
```
